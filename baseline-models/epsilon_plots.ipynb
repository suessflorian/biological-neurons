```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('csvs/epsilon_plots.csv')
data
```

```python
model = 'LeNet5'
dataset = 'MNIST'

subset = data[(data.model == model) & (data.dataset == dataset)]

for attack in ['LinfFastGradientAttack', 'LInfFMNAttack', 'LinfDeepFoolAttack', 'SquareAttack']:
    subset_attack = subset[subset.attack == attack]
    plt.plot(subset_attack.epsilon, subset_attack.susceptibility_rate)

plt.xlabel('Epsilon')
plt.ylabel('Attack Success Rate')
plt.legend(['FGSM', 'FMNA', 'DeepFool', 'Square'])
plt.title(f'Attacks on {model}, {dataset} Dataset')
low, high = plt.ylim()
plt.vlines(subset_attack.epsilon, ymin=max(0, low), ymax=min(1, high), color='black', alpha=0.25, linestyle='dashed')
plt.show()
```
